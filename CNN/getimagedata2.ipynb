```python
from PIL import Image
import requests
import io
from google.colab import files
from io import BytesIO
```

```python
# url da imagem
uploaded = files.upload()
if uploaded:
  for nome_arquivo in uploaded.keys():
      print(nome_arquivo)
      image = Image.open(io.BytesIO(uploaded[nome_arquivo])).convert("L").resize((28, 28))
      image = Image.open(BytesIO(uploaded[nome_arquivo])).resize((28, 28))
      image.show()
      break
```

```python
# Usa getpixel() para acessar um pixel (ex: coordenada x=10, y=20)
x, y = 0, 0
pixel = image.getpixel((x, y))
if pixel !=0:
  print("Valores RGB do pixel em (",x,", ",y,"):", pixel)
  print("R:", pixel[0])
  print("G:", pixel[1])
  print("B:", pixel[2])
```

```python
# pegando os pixels da imagem inteira
for x in range(image.width):
    for y in range(image.height):
        pixel = image.getpixel((x, y))
        print("Valores RGB do pixel em (",x,", ",y,"):", pixel)
```
